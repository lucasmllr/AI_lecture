```python
from kmeans import *
import pylab as plt
from scipy.ndimage import imread
```

# 10.1) k-means for image compression

```python
#loading data
im = imread('grass2.jpg')

plt.imshow(im)
plt.title('original image')
plt.show()
```

```python
#compression with two means

k2 = compress_RGB(im, k=2)

plt.imshow(k2)
plt.title('compression with two clusters')
plt.show()
```

```python
#compression with 5 means

k5 = compress_RGB(im, k=5)

plt.imshow(k5)
plt.title('compression with five clusters')
plt.show()
```

```python
#compression with 10 means

k10 = compress_RGB(im, k=10)

plt.imshow(k10)
plt.title('compression with ten clusters')
plt.show()
```

```python
#compression with 100 clusters

k100 = compress_RGB(im, k=100)

plt.imshow(k100)
plt.title('compression with 100 clusters')
plt.show()
```

For reasonably good image quality that is close to the colors of the original image one would probably need to fit a few thousand clusters. The scipy k-means algorithm used is, however, very slow and already takes long for 100 clusters.
