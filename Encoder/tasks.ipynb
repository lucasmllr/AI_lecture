# 9.1 Variants of SGD

## batch gradient descent

In batch gradient descent an optimization step is done only after having computed a loss for all instances inside the whole batch. The gradient w.r.t the parameters is calculated on the sum of all individual losses.

## mini-batch gradient descent

In mini-batch gradient descent the gradient is not computed on the sum over all losses in a batch but only a certain fraction.

## stochastic gradient descent

In SGD a gradient is calculated on every individual loss of a single instance.
